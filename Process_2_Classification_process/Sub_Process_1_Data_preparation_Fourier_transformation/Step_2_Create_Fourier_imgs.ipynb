#### Step_1_Loop through each year_range to create the harmonic imgs

```python
for start,end in year_range:
    
    # define the year_name
    year_name = f'{start[:4]}_{end[:4]}'

    for region_CN, region_EN in NAME_CN_EN_dict.items():

        # construct the export name    
        export_name = f'Fourier_{region_EN}_{year}'

        # create the export img/region
        out_region = Target_area.filter(ee.Filter.equals('NAME1',region_CN))

        # ============== Make Fourier img ==============
    
        # instatiate the instance
        Fourier_instance = Make_Fourier(start_date=start,
                                        end_date=end,
                                        area=out_region)
        
        # make result imgs
        Fourier_instance.Stp_1_Create_hamonic_names()
        Fourier_instance.Stp_2_Add_harmonics()
        Fourier_instance.Stp_3_Harmonic_fit()
        
        # Fourier img
        img = Fourier_instance.harmonicTrendCoefficients
        out_img = img.clip(out_region)
    
        # ================= Export ==============
        
        # construct the export task
        task = ee.batch.Export.image.toAsset(image= out_img,
                                             description=export_name,
                                             assetId=f'{path}/{export_name}',
                                             region=out_region.geometry().bounds(),
                                             scale=30,
                                             maxPixels=int(1e13))
    
        task.start()
    
        # print out the process
        print(export_name)
```
